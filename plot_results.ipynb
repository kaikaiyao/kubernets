```python
import matplotlib.pyplot as plt

# Data
delta_vals = [0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1, 2]
# Create equally spaced x positions for categorical ticks
x = range(len(delta_vals))

auc   = [0.6006, 0.6030, 0.9736, 0.8454, 0.9618, 1.0000, 1.0000, 0.9999]
tpr   = [0.0210, 0.0268, 0.6665, 0.0934, 0.6712, 1.0000, 1.0000, 0.9999]
lpips = [0.0048, 0.0111, 0.0555, 0.0600, 0.1347, 0.2876, 0.3452, 0.3984]
fid   = [0.1825, 0.3464, 3.2328, 3.1628, 8.0560, 24.9875, 35.5544, 53.6218]

# Create primary axis for AUC, TPR, and LPIPS
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot AUC and TPR with blue theme on left y-axis
l1, = ax1.plot(x, auc, marker='o', color='blue', label='AUC')
l2, = ax1.plot(x, tpr, marker='o', color='deepskyblue', label='TPR@1%FPR')
# Plot LPIPS with orange theme on left y-axis
l3, = ax1.plot(x, lpips, marker='o', color='darkorange', label='LPIPS')

ax1.set_xlabel('Delta')
ax1.set_ylabel('AUC / TPR@1%FPR / LPIPS', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Create secondary axis for FID using an orange theme
ax2 = ax1.twinx()
l4, = ax2.plot(x, fid, marker='o', color='orange', label='FID')
ax2.set_ylabel('FID', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Set x-axis ticks to equally spaced positions and label them with delta values
ax1.set_xticks(x)
ax1.set_xticklabels(delta_vals)
plt.title('Ablation Study on Delta')

# Combine legends from both axes
lines = [l1, l2, l3, l4]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

plt.tight_layout()
plt.show()
```
